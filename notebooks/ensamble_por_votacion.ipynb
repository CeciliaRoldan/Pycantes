```python
#pandas libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


%run "../notebooks/cargar_df.ipynb"
```

```python
prediccion_random_forest = pd.read_csv("prediccion_random_forest")
prediccion_xgboost = pd.read_csv("prediccion_xgboost_6_categorical")
prediccion_knn = pd.read_csv("prediccion_knn")
```

```python
prediccion_random_forest.head()
```

```python
rf_xgb = prediccion_random_forest.merge(prediccion_xgboost, how='inner', on='Opportunity_ID')
rf_xgb.head()
```

```python
ensamble = rf_xgb.merge(prediccion_knn, how='inner', on='Opportunity_ID')
ensamble.head()
```

```python
col = ensamble.loc[: , "Target_x":"Target"]
ensamble['salary_mean'] = col.mean(axis=1)
```

```python
ensamble.head()
```

```python
ensamble['ensamble_target'] = ensamble['salary_mean'].apply(lambda x: 1 if x >= 0.5 else 0)
```

```python
ensamble.head()
```

```python
prediccion = ensamble[["Opportunity_ID", "ensamble_target"]].rename(columns={"ensamble_target": "target"})
prediccion
```

```python

prediccion.to_csv('prediccion_ensamble', index=False)
```
